## Persiapan Data

Mengimpor modul yang diperlukan

```
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
```

Mengunduh dataset COVID-19 seluruh negara di dunia

```
confirmed_cases_file_link='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
death_cases_file_link='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
recovered_cases_file_link='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
```

```
confirmed_df=pd.read_csv(confirmed_cases_file_link)
deaths_df=pd.read_csv(death_cases_file_link)
recovered_df=pd.read_csv(recovered_cases_file_link)
```

Membersihkan dataset

```
confirmed_df=confirmed_df.replace(np.nan,'',regex=True)
deaths_df=deaths_df.replace(np.nan, '', regex=True)
recovered_df=recovered_df.replace(np.nan, '', regex=True)
```

```
confirmed_df=confirmed_df.copy().drop(['Lat', 'Long','Province/State'],axis=1)
deaths_df=deaths_df.copy().drop(['Lat','Long','Province/State'],axis=1)
recovered_df=recovered_df.copy().drop(['Lat','Long','Province/State'],axis=1)
```

## Diagram Garis & Batang Stack COVID-19 Indonesia

Mengambil data COVID-19 Indonesia

```
indo_conf_df=confirmed_df[confirmed_df['Country/Region']=='Indonesia']
indo_death_df=deaths_df[deaths_df['Country/Region']=='Indonesia']
indo_recov_df=recovered_df[recovered_df['Country/Region']=='Indonesia']
```

```
indo_conf=indo_conf_df.transpose()
indo_death=indo_death_df.transpose()
indo_recov=indo_recov_df.transpose()
```

```
indo_conf=indo_conf[1::]
indo_conf.columns=['Conf']
indo_death=indo_death[1::]
indo_death.columns=['Death']
indo_recov=indo_recov[1::]
indo_recov.columns=['Recov']
```

Ploting diagram garis COVID-19 Indonesia

```
gbr_1=go.Figure()
gbr_1.add_trace(go.Scatter(x=indo_conf.index,y=indo_conf['Conf'], name='Konfirmasi',mode='lines'))
gbr_1.add_trace(go.Scatter(x=indo_death.index,y=indo_death['Death'], name='Meninggal',mode='lines'))
gbr_1.add_trace(go.Scatter(x=indo_recov.index,y=indo_recov['Recov'], name='Sembuh',mode='lines'))
gbr_1.update_layout(title='Diagram Garis COVID-19 Indonesia Hingga 29 Juni 2020', yaxis_title='Jumlah', xaxis_title='Tanggal')
gbr_1.show()
```

Ploting diagram batang stack COVID-19 Indonesia

```
index=indo_conf.index
gbr_2=go.Figure(data=[
                      go.Bar(name='Konfirmasi',x=index, y=indo_conf['Conf']),
                      go.Bar(name='Meninggal', x=index, y=indo_death['Death']),
                      go.Bar(name='Sembuh', x=index, y=indo_recov['Recov'])
                      
])

gbr_2.update_layout(barmode='stack',title='Diagram Batang Stack COVID-19 Indonesia Hingga 29 Juni 2020', yaxis_title='Jumlah', xaxis_title='Tanggal')
gbr_2.show()
```

## Diagram Garis Data Harian COVID-19 Indonesia

Menyiapkan dataset kenaikan

```
kenaikan_conf = []
for i in range(len(indo_conf.Conf)-1):
  kenaikan_conf.append(indo_conf.Conf[i+1]-indo_conf.Conf[i])

kenaikan_death = []
for i in range(len(indo_death.Death)-1):
  kenaikan_death.append(indo_death.Death[i+1]-indo_death.Death[i])

kenaikan_recov = []
for i in range(len(indo_recov.Recov)-1):
  kenaikan_recov.append(indo_recov.Recov[i+1]-indo_recov.Recov[i])
```

Ploting Diagram Garis Data Harian COVID-19 Indonesia

```
index=indo_conf.index
gbr_3=go.Figure()
gbr_3.add_trace(go.Scatter(x=index,y=kenaikan_conf, name='Konfirmasi',mode='lines+markers'))
gbr_3.add_trace(go.Scatter(x=index,y=kenaikan_death, name='Meninggal',mode='lines+markers'))
gbr_3.add_trace(go.Scatter(x=index,y=kenaikan_recov, name='Sembuh',mode='lines+markers'))
gbr_3.update_layout(title='Diagram Garis Harian COVID-19 Indonesia Hingga 29 Juni 2020', yaxis_title='Jumlah', xaxis_title='Tanggal')
gbr_3.show()
```

## COVID-19 di Negara-negara ASEAN
